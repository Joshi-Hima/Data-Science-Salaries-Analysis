```python
#importing the important libraries that would be needed

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Q1) 

```python
dframe = pd.read_csv('DataScienceSalaries.csv') #reads the data from the csv file uploaded
dframe  #prints the data frame in table 
```

# Q2)

```python
#Removing columns: salary and salary_currency

dframe_removed_columns = dframe.drop(columns=['salary','salary_currency'])
dframe_removed_columns
```

# Q3)

```python
#Removing the NaN missing values from the updated dataframe

dframe_removed_missing_values = dframe_removed_columns.dropna()
dframe_removed_missing_values
```

# Q4)

```python
# Checking the duplicate values in the dataframe

dframe[dframe.duplicated()]
```

# Q5)

```python
#finding unique values in each column in the dataframe
#using for loop to iterate through each column

for each_col in dframe.columns:
    unique_in_each_col = dframe[each_col].unique()
    print(f"The unique values in {each_col} are {unique_in_each_col}")
```

# Q6)

```python
#Renaming the experience_level column values as per question

dframe['experience_level'] = dframe['experience_level'].replace({
    'SE': 'Senior Level/Expert',
    'MI': 'Medium Level/Intermediate',
    'EN': 'Entry Level',
    'EX': 'Executive Level'
})
dframe
```

# Data Analysis

# Q7)

```python
# To show the summary statics of 'salary' column
from scipy.stats import skew, kurtosis

sum_salary = sum(dframe['salary'])
print(f"Sum is: {sum_salary}")

mean_salary = dframe['salary'].mean()
print(f"Mean is: {mean_salary}")

std_salary = dframe['salary'].std()
print(f"Standard Deviation is: {std_salary}")

sekweness_salary = skew(dframe['salary'])
print(f"Skewness is: {sekweness_salary}")

kurtosis_salary = kurtosis(dframe['salary'])
print(f"Kurtosis is: {kurtosis_salary}")
```

# Q8)

```python
#finding the correlation between different variables

#firstly converting all into numeric as correlation accepts only numeric values

string_to_numeric = dframe.select_dtypes(include=['int', 'float']).columns
dframe[string_to_numeric] = dframe[string_to_numeric].apply(pd.to_numeric, errors='coerce')

#now calculating the correlation of it
correlation_variables = dframe[string_to_numeric].corr()
correlation_variables
```

# Data Exploration

# Q9)

```python
# Finding out top 15 jobs

#First fiding the number of repetation of jobs in 'job_title' column

no_of_each_job = dframe['job_title'].value_counts()

#Now using head function to find out the top 15 of the jobs

top_15_jobs_according_to_repetation = no_of_each_job.head(15)
print("The top 15 jobs according to repatation are: ")
print(top_15_jobs_according_to_repetation)

#Now plotting it in bar graph

top_15_jobs_according_to_repetation.plot(kind='bar', color='lightcoral')
plt.title('Top 15 Jobs according to repetation', size=20)
plt.xlabel('Job Title', size=25, labelpad=15)
plt.ylabel('No. of repetations', size=25, labelpad=10)
plt.show()
```

# Q10)

```python
# Top 15 jobs having the highest salaries

#Firstly grouping the 'job_title' with 'salary_in_usd'

group_job_title_with_salaries = dframe.groupby('job_title')['salary_in_usd'].sum().reset_index()

#Now sorting the values by 'salary_in_usd'

top_15_jobs_by_acc_salaries = group_job_title_with_salaries.sort_values(by='salary_in_usd').head(15)

#Now plotting it in bar graph

top_15_jobs_by_acc_salaries.plot(kind='bar', x='job_title', y='salary_in_usd', color='lightcoral')
plt.title('Top 15 Jobs According to Highest Salaries', size=20)
plt.xlabel('Top 15 Jobs', size=25, labelpad=25)
plt.ylabel('Salary in USD', size=25, labelpad=15)
plt.show()
```

# Q11)

```python
#Finding salaries baed on experience level

#Firstly grouping experience level with 'salary_in_usd'

salary_on_exp_level = dframe.groupby('experience_level')['salary_in_usd'].mean()

#Now plotting it in bar graph

salary_on_exp_level.plot(kind='bar', x='experience_level', y='salary_in_usd', color='lightcoral')
plt.title('Salaries based on Experience Level', size=20)
plt.xlabel('Experience Level', size=25, labelpad=25)
plt.ylabel('Salary in USD', size=25, labelpad=15)
plt.show()
```

# Q12)

```python
#Making Histogram of 'salary_in_usd'

plt.hist(dframe['salary_in_usd'], color='lightcoral')
plt.title('Histogram of Salary in USD', size=20)
plt.xlabel('Salary in USD', size=25, labelpad=15)
plt.ylabel('Frequency of the Salary', size=20, labelpad=10)
```

```python
#Making the box plot of 'remote_ratio'

plt.boxplot(dframe['remote_ratio'])
plt.title('Box Plot of Remote Ratio',size=25)
plt.xlabel('Remote Ratio', size=20, labelpad=15)
plt.ylabel('Values of Remote Ratio', size=20, labelpad=10)
plt.show()
```
